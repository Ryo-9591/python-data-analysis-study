## 描画オブジェクト

```python
import matplotlib.pyplot as plt
import matplotlib.style

# ggplotスタイルを指定
matplotlib.style.use("ggplot")
```

### 描画オブジェクトとサブプロット

```python
fig, axes = plt.subplots(2)  # 2つのサブプロットを配置
plt.show()
```

```python
fig, axes = plt.subplots(2, 2)  # 2行2列のサブプロットを配置
plt.show()
```

### タイトル・軸ラベル

```python
# 1行2列のサブプロットを配置
fig, axes = plt.subplots(ncols=2)

# サブプロットにタイトルを設定
axes[0].set_title("subplot title 0")  
axes[0].set_xlabel("x label")  # X軸にラベルを設定
axes[0].set_ylabel("y label")  # Y軸にラベルを設定
axes[1].set_title("subplot title 1")
# 描画オブジェクトにタイトルを設定
fig.suptitle("figure title")

plt.show()
```

```python
fig, ax = plt.subplots()

# 凡例用のラベルを設定
ax.plot([1, 2, 3], [2, 4, 9], label="legend label")
ax.legend(loc="best")  # 凡例を表示

plt.show()
```
